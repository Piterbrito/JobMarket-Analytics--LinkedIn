```python
# Splinter bot
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
browser.visit(url)

print("INSERT The Variables Values AS STRING with NO space")


def scrapperMaster(city, position): #INSERT The Variables Values AS STRING with NO space

    # Loop trhu job linting Pages
    for start in range(1, 5, 1):
        
        url = f"https://www.linkedin.com/jobs/search/?keywords={position}&location={city}&locationId=us%3A84&start={str(start*25)}"
        
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        
        browser.visit(url)
        html = browser.html
        soup = BeautifulSoup(response.text, 'html.parser')

        # Retriving all target data 
        
        title = soup.find_all("h3")
        company = soup.find_all("h4")
        
        paragraph = soup.find_all('p')
        lines = soup.find_all('li')

        # Extracting the text and saving in lists
        
        for t in title:
            #print(t.text)
            analyst_title.append(t.text)    
        for c in company:
            #print(t.text)
            analyst_company.append(c.text)
            
            time.sleep(1)  
        
            for p in paragraph:
                #print(p.text)
                p_tags.append(p.text) # Alert! do not print, a lot records, it mid burn your memory!
                
            # after get all <p> then store as array     
            analyst_article.append(p_tags) 
            
            # raw data
            for l in lines:
                print(l.text)
                analyst_lines.append(l.text)            

        
        button = browser.find_by_class('class="result-card job-result-card result-card--with-hover-state"')
        button.click()
        
        time.sleep(10)   
        
        
    return dic
        
    
    
 # Calling the Function   
scrapperMaster("LosAngeles", "Analyst")
```
